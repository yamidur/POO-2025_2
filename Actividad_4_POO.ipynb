La clase contiene un constructor que inicialice los atributos de la clase. Adem√°s, la clase posee los siguientes m√©todos:

- Imprimir: muestra por pantalla los valores de sus atributos.

- Verificar edad: este m√©todo recibe como par√°metro un valor entero que representa la edad del vendedor. Para que un vendedor pueda desempe√±ar sus labores se requiere que sea mayor de edad (mayor de 18 a√±os). Si esta condici√≥n no se cumple, se lanza una excepci√≥n de tipo IllegalArgumentException con el mensaje ‚ÄúEl vendedor debe ser mayor de 18 a√±os‚Äù. Adem√°s, se eval√∫a si la edad

una excepci√≥n de tipo IllegalArgumentException con el mensaje ‚ÄúLa

```python
# @title
import ipywidgets as widgets
from IPython.display import display, HTML
import re # Usado para validaci√≥n simple de texto

# -----------------------------------------------------------------
# 1. Definici√≥n de la L√≥gica (Clase)
# -----------------------------------------------------------------

class VendedorAppLogic:
    """Clase que maneja la l√≥gica de validaci√≥n de los datos del vendedor."""
    def __init__(self, nombre, apellidos, edad):
        self.nombre = nombre
        self.apellidos = apellidos
        self.edad = edad

    def validar_y_obtener_info(self):
        """Valida la edad y devuelve la informaci√≥n formateada o levanta ValueError."""
        try:
            edad_valor = int(self.edad)
        except ValueError:
            raise ValueError("La edad debe ser un n√∫mero entero.")

        if edad_valor < 18:
            raise ValueError("El vendedor debe ser mayor de 18 a√±os.")

        # Asumiendo que el nombre y apellido no deben estar vac√≠os
        if not self.nombre.strip() or not self.apellidos.strip():
            raise ValueError("Nombre y Apellidos no pueden estar vac√≠os.")

        if not (0 <= edad_valor <= 120):
            raise ValueError("La edad debe estar entre 0 y 120 a√±os.")

        return (
            f"Nombre: {self.nombre}\n"
            f"Apellidos: {self.apellidos}\n"
            f"Edad: {edad_valor} a√±os"
        )

# -----------------------------------------------------------------
# 2. Interfaz Gr√°fica (Widgets)
# -----------------------------------------------------------------

# Definici√≥n de estilos
layout_ancho = widgets.Layout(width='200px')
estilo_label = widgets.Layout(width='100px', justify_content='flex-start')

# Widgets de entrada
titulo = widgets.HTML("<h2>Registro de Vendedor</h2>")

label_nombre = widgets.Label(value="Nombre:", layout=estilo_label)
input_nombre = widgets.Text(description="", layout=layout_ancho)
box_nombre = widgets.HBox([label_nombre, input_nombre])

label_apellidos = widgets.Label(value="Apellidos:", layout=estilo_label)
input_apellidos = widgets.Text(description="", layout=layout_ancho)
box_apellidos = widgets.HBox([label_apellidos, input_apellidos])

label_edad = widgets.Label(value="Edad:", layout=estilo_label)
input_edad = widgets.Text(description="", layout=layout_ancho) # Usamos Text para manejar la validaci√≥n en el c√≥digo
box_edad = widgets.HBox([label_edad, input_edad])

# Bot√≥n y Salida
boton_registrar = widgets.Button(description='Registrar', button_style='success', layout=widgets.Layout(width='150px'))
output_mensaje = widgets.Output()


# -----------------------------------------------------------------
# 3. Funci√≥n de Manejo de Eventos (Callback)
# -----------------------------------------------------------------

def on_registrar_clicked(b):
    """Funci√≥n que se ejecuta al hacer clic en el bot√≥n 'Registrar'."""
    with output_mensaje:
        output_mensaje.clear_output()

        # 1. Recuperar valores
        nombre = input_nombre.value
        apellidos = input_apellidos.value
        edad_str = input_edad.value

        try:
            # 2. Ejecutar la l√≥gica de validaci√≥n
            vendedor = VendedorAppLogic(nombre, apellidos, edad_str)
            info = vendedor.validar_y_obtener_info()

            # 3. Mostrar resultado exitoso (Simula messagebox.showinfo)
            print("--- REGISTRO EXITOSO ---")
            print(info)

        except ValueError as e:
            # 4. Mostrar error (Simula messagebox.showerror)
            display(HTML(f'<p style="color: red; font-weight: bold;">ERROR: {e}</p>'))

# Asociar el evento del bot√≥n a la funci√≥n
boton_registrar.on_click(on_registrar_clicked)


# -----------------------------------------------------------------
# 4. Despliegue de la Interfaz
# -----------------------------------------------------------------

interfaz_completa = widgets.VBox([
    titulo,
    box_nombre,
    box_apellidos,
    box_edad,
    boton_registrar,
    widgets.HBox([widgets.Label(value="--- Resultados ---", layout=widgets.Layout(width='500px'))]),
    output_mensaje
], layout=widgets.Layout(border='1px solid lightgray', padding='20px', width='450px'))

display(interfaz_completa)
```

```python
# @title
import ipywidgets as widgets
from IPython.display import display, HTML
import math
import re # Usado para validaci√≥n

# ============================================================
# CLASE DE L√ìGICA (CalculosApp adaptada)
# ============================================================

class CalculadoraNumerica:
    """Clase para encapsular la l√≥gica de c√°lculo de logaritmo y ra√≠z cuadrada."""

    def __init__(self, valor):
        self.valor = valor

    def calcular_logaritmo(self):
        """Calcula el logaritmo natural."""
        return math.log(self.valor)

    def calcular_raiz_cuadrada(self):
        """Calcula la ra√≠z cuadrada."""
        return math.sqrt(self.valor)

# ============================================================
# L√ìGICA DE LA INTERFAZ (Reemplazando Tkinter con ipywidgets)
# ============================================================

# Widgets de Interfaz
output_area = widgets.Output()
valor_entry = widgets.Text(
    description='Ingrese un n√∫mero positivo:',
    placeholder='Ej: 25.5',
    layout=widgets.Layout(width='300px')
)
btn_calcular = widgets.Button(
    description="Calcular",
    button_style='success',
    layout=widgets.Layout(width='120px')
)

# √Årea para mostrar resultados
resultado_ln = widgets.Label(value="")
resultado_sqrt = widgets.Label(value="")

def calcular_valores(btn):
    """Funci√≥n llamada al presionar el bot√≥n 'Calcular'."""
    with output_area:
        output_area.clear_output()
        errores = []
        valor = None

        try:
            valor_str = valor_entry.value.strip()

            # 1. Validaci√≥n de entrada num√©rica
            if not re.match(r'^-?\d+(\.\d+)?$', valor_str):
                 raise ValueError("El valor debe ser num√©rico.")

            valor = float(valor_str)

            # 2. Verificaci√≥n de valor negativo (para los c√°lculos espec√≠ficos)
            if valor < 0:
                errores.append("Error: El valor debe ser un n√∫mero positivo para calcular el logaritmo.")
                errores.append("Error: El valor debe ser un n√∫mero positivo para calcular la ra√≠z cuadrada.")
                resultado_ln.value = "ln() = N/A"
                resultado_sqrt.value = "‚àö() = N/A"
            else:
                # 3. Realizar c√°lculos
                calculadora = CalculadoraNumerica(valor)
                ln_val = calculadora.calcular_logaritmo()
                sqrt_val = calculadora.calcular_raiz_cuadrada()

                # Mostrar resultados
                resultado_ln.value = f"ln({valor}) = {ln_val:.4f}"
                resultado_sqrt.value = f"‚àö{valor} = {sqrt_val:.4f}"

        except ValueError as e:
            # Capturar errores de conversi√≥n no num√©rica
            errores.append(f"Error: {e}")
            resultado_ln.value = "ln() = N/A"
            resultado_sqrt.value = "‚àö() = N/A"

        except Exception as e:
            # Capturar otros errores (como log(0) o math.sqrt(negativo) si no se usa el if valor < 0)
            if "math domain error" in str(e):
                 errores.append("Error: El logaritmo de cero y la ra√≠z de un negativo no est√°n definidos en los reales.")
            else:
                errores.append(f"Error inesperado: {e}")
            resultado_ln.value = "ln() = N/A"
            resultado_sqrt.value = "‚àö() = N/A"

        # Mostrar errores si los hay
        if errores:
            # Reemplazamos messagebox.showerror con una visualizaci√≥n HTML de error
            display(HTML(f'<div style="color: red; font-weight: bold; padding: 5px;">Errores encontrados:<br>- {"<br>- ".join(errores)}</div>'))
        elif valor is not None and valor >= 0:
             display(HTML(f'<div style="color: green; font-weight: bold;">‚úÖ C√°lculo completado.</div>'))


# Asignar la funci√≥n al bot√≥n
btn_calcular.on_click(calcular_valores)

# Contenedor principal
main_container = widgets.VBox([
    widgets.HTML("<h2>C√°lculos Num√©ricos (Logaritmo y Ra√≠z Cuadrada)</h2>"),
    widgets.HBox([
        widgets.Label("Ingrese un n√∫mero positivo:"),
        valor_entry
    ], layout=widgets.Layout(align_items='center', margin='10px 0')),

    btn_calcular,
    widgets.HTML("<hr style='width: 100%;'>"),

    widgets.VBox([
        widgets.HTML("<h3>Resultados:</h3>"),
        resultado_ln,
        resultado_sqrt
    ], layout=widgets.Layout(align_items='flex-start', border='1px solid #ddd', padding='15px', width='90%')),

    output_area
], layout=widgets.Layout(align_items='center', border='1px solid #ccc', padding='20px', width='500px'))

# Mostrar la interfaz
display(main_container)
```

```python
# @title

import ipywidgets as widgets
from IPython.display import display, HTML

# -----------------------------
# Clases
# -----------------------------
class Programador:
    def __init__(self, nombre, apellidos):
        self.nombre = nombre
        self.apellidos = apellidos

class EquipoMaratonProgramacion:
    def __init__(self, nombre_equipo, universidad, lenguaje_programacion):
        self.nombre_equipo = nombre_equipo
        self.universidad = universidad
        self.lenguaje_programacion = lenguaje_programacion
        self.programadores = []

    def anadir_programador(self, programador):
        if len(self.programadores) >= 3:
            raise Exception("El equipo no puede tener m√°s de 3 programadores.")
        self.programadores.append(programador)

    @staticmethod
    def validar_campo(texto):
        if any(char.isdigit() for char in texto):
            raise Exception("El campo no puede contener d√≠gitos.")
        if len(texto) > 20:
            raise Exception("El campo no debe tener m√°s de 20 caracteres.")

# -----------------------------
# Widgets (interfaz Colab)
# -----------------------------
entry_nombre_equipo = widgets.Text(description="Equipo:", layout=widgets.Layout(width='400px'))
entry_universidad = widgets.Text(description="Universidad:", layout=widgets.Layout(width='400px'))
entry_lenguaje = widgets.Text(description="Lenguaje:", layout=widgets.Layout(width='400px'))

entries_nombres = []
entries_apellidos = []

# Crear campos para 3 integrantes
integrantes_widgets = []
for i in range(3):
    lbl_integrante = widgets.HTML(f"<h4>üë®‚Äçüíª Integrante {i+1}</h4>")
    entry_nombre = widgets.Text(description="Nombre:", layout=widgets.Layout(width='350px'))
    entry_apellido = widgets.Text(description="Apellidos:", layout=widgets.Layout(width='350px'))
    entries_nombres.append(entry_nombre)
    entries_apellidos.append(entry_apellido)
    integrantes_widgets.append(widgets.VBox([lbl_integrante, entry_nombre, entry_apellido]))

# Bot√≥n y salida
btn_registrar = widgets.Button(description="Registrar equipo", button_style='success')
output = widgets.Output()

# -----------------------------
# L√≥gica de registro
# -----------------------------
def registrar_equipo(b):
    with output:
        output.clear_output()
        try:
            nombre_equipo = entry_nombre_equipo.value.strip()
            universidad = entry_universidad.value.strip()
            lenguaje = entry_lenguaje.value.strip()

            if not nombre_equipo or not universidad or not lenguaje:
                raise Exception("Todos los campos del equipo son obligatorios.")

            equipo = EquipoMaratonProgramacion(nombre_equipo, universidad, lenguaje)

            programadores_validos = 0
            for i in range(3):
                nombre = entries_nombres[i].value.strip()
                apellidos = entries_apellidos[i].value.strip()

                if nombre and apellidos:
                    EquipoMaratonProgramacion.validar_campo(nombre)
                    EquipoMaratonProgramacion.validar_campo(apellidos)
                    equipo.anadir_programador(Programador(nombre, apellidos))
                    programadores_validos += 1
                elif nombre or apellidos:
                    raise Exception(f"Debes completar nombre y apellidos del integrante {i+1} o dejar ambos en blanco.")

            if programadores_validos < 2:
                raise Exception("Debes registrar al menos 2 programadores.")

            print(f"‚úÖ Equipo '{equipo.nombre_equipo}' registrado correctamente con {programadores_validos} integrantes.")
        except Exception as e:
            print(f"‚ùå Error: {e}")

btn_registrar.on_click(registrar_equipo)

# -----------------------------
# Mostrar interfaz
# -----------------------------
display(HTML("<h3 style='color:#1a5276'>Ejercicio 6.7 - Registro de Equipo (m√≠nimo 2 integrantes)</h3>"))
display(HTML("<b>Datos del Equipo</b>"))
display(entry_nombre_equipo, entry_universidad, entry_lenguaje)
display(HTML("<hr><b>Datos de los Programadores (m√≠nimo 2)</b>"))

for box in integrantes_widgets:
    display(box)

display(btn_registrar, output)
```

```python
# @title
# ============================================================
# DIAGRAMA UML DE CLASES: M√ìDULO ARCHIVOS (POSICI√ìN DE PESTA√ëA CORREGIDA)
# ============================================================

!apt-get install -y graphviz > /dev/null 2>&1
!pip install graphviz > /dev/null 2>&1

from graphviz import Digraph
from IPython.display import Image, display

# Crear el diagrama base
uml = Digraph('DiagramaArchivosCorregido2', filename='diagrama_archivos_corregido2', format='png')

# --------------------------------------------------------------
# Configuraci√≥n general y estilo
# --------------------------------------------------------------
uml.attr(
    rankdir='TB', # Top to Bottom (general)
    fontsize='12',
    fontname='Helvetica',
    ranksep='0.6',
    nodesep='0.4'
)
uml.attr('graph', dpi='100')

# Estilo por defecto para los nodos de clase (usando RECORD estable)
uml.attr('node',
          shape='record',
          style='filled',
          fillcolor='white',
          color='black',
          fontname='Helvetica',
          fontsize='10',
          margin='0.1')

# --------------------------------------------------------------
# Paquete Archivos (Cluster y Pesta√±a)
# --------------------------------------------------------------
with uml.subgraph(name='cluster_Archivos') as c:
    # --- Estilo del Cluster (Contenedor del Paquete) ---
    c.attr(style='rounded,filled', fillcolor='gray95', color='black', label='')

    # --- 1. Nodo Pesta√±a (T√≠tulo "Archivos") ---
    c.node('package_tab',
           'Archivos',
           shape='tab',
           style='filled',
           fillcolor='white',
           color='black',
           fontname='Helvetica',
           fontsize='12',
           # Ajustamos el margen para que parezca m√°s compacto
           margin='0.05')

    # --- 2. Definiciones de Clases ---
    c.attr('node', fillcolor='#F0F0E0', margin='0.15') # Color de fondo claro para las clases

    # Clase LeerArchivo (Arriba)
    c.node('LeerArchivo',
           '{(C) LeerArchivo | \u2022 leer_archivo(): void}',
           height='0.5')

    # Clase open (Medio Izquierda)
    c.node('open', '{(C) open | }')

    # Clase ScrolledText (Medio Derecha)
    c.node('ScrolledText', '{(C) ScrolledText | }')

    # Clase TextIOWrapper (Abajo)
    c.node('TextIOWrapper', '{(C) TextIOWrapper | }')

    # --- 3. Relaciones ---

    c.attr('edge', arrowhead='normal', style='solid', fontsize='9')

    # LeerArchivo -> open (usa)
    c.edge('LeerArchivo', 'open', label='usa')

    # LeerArchivo -> ScrolledText (muestra resultado)
    c.edge('LeerArchivo', 'ScrolledText', label='muestra resultado')

    # open -> TextIOWrapper (devuelve)
    c.edge('open', 'TextIOWrapper', label='devuelve')

    # --- 4. Control de Posici√≥n (CLAVE PARA LA PESTA√ëA Y LA ESTRUCTURA) ---

    # a) Forzar la estructura vertical (LeerArchivo -> open/ScrolledText -> TextIOWrapper)
    c.edge('LeerArchivo', 'open', style='invis') # Define el flujo principal
    c.edge('open', 'TextIOWrapper', style='invis')

    # b) Forzar 'open' y 'ScrolledText' a estar en el mismo nivel horizontal
    c.attr(rank='same')
    c.node('open')
    c.node('ScrolledText')

    # c) CONEXI√ìN DE LA PESTA√ëA: Conectar el nodo de la pesta√±a a la primera clase
    #    Esto asegura que la pesta√±a est√© justo encima de la primera columna
    c.edge('package_tab', 'LeerArchivo', style='invis', weight='100')
    # Usamos un peso alto para asegurar que la pesta√±a quede en el nivel superior.


# Generar y mostrar el diagrama
uml.render(cleanup=True)
display(Image(filename='diagrama_archivos_corregido2.png'))
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML
import io
from google.colab import files # Necesario para la subida de archivos en Colab

# ============================================================
# LECTOR DE ARCHIVOS PARA COLAB
# ============================================================

# Widgets de Interfaz
output_area = widgets.Output()
text_area_content = widgets.Textarea(
    value='',
    placeholder='El contenido del archivo se mostrar√° aqu√≠...',
    description='Contenido:',
    disabled=False,
    layout=widgets.Layout(width='95%', height='300px')
)
btn_abrir = widgets.Button(
    description="Abrir archivo .txt",
    button_style='primary',
    tooltip='Sube un archivo .txt',
    layout=widgets.Layout(width='200px')
)

def leer_archivo_colab(btn):
    """
    Funci√≥n que simula filedialog.askopenfilename de Tkinter usando
    el m√≥dulo 'files' de Google Colab para la subida.
    """
    text_area_content.value = 'Cargando archivo...'
    with output_area:
        output_area.clear_output()

        try:
            # Sube un archivo. El usuario seleccionar√° el archivo en un di√°logo
            uploaded = files.upload()
        except Exception:
            # Esto ocurre si el usuario cancela la selecci√≥n
            text_area_content.value = 'Selecci√≥n de archivo cancelada.'
            return

        # Procesar el archivo subido
        if not uploaded:
            text_area_content.value = 'Selecci√≥n de archivo cancelada.'
            return

        # Tomar el nombre del primer archivo subido
        nombre_archivo = list(uploaded.keys())[0]

        # Verificar si es un archivo .txt
        if not nombre_archivo.lower().endswith('.txt'):
            display(HTML('<div style="color: red; font-weight: bold;">‚ùå Error: Por favor, seleccione un archivo con extensi√≥n .txt.</div>'))
            text_area_content.value = 'Error: Archivo no compatible.'
            return

        try:
            # Leer el contenido del archivo subido
            contenido = uploaded[nombre_archivo].decode('utf-8')

            # Mostrar el contenido en el Textarea
            text_area_content.value = contenido
            display(HTML(f'<div style="color: green;">‚úÖ Archivo "{nombre_archivo}" cargado exitosamente.</div>'))

        except Exception as e:
            display(HTML(f'<div style="color: red; font-weight: bold;">‚ùå Error de lectura: No se pudo leer el archivo. ({e})</div>'))
            text_area_content.value = f"Error al procesar el archivo: {e}"

# Asignar la funci√≥n al bot√≥n
btn_abrir.on_click(leer_archivo_colab)

# Contenedor principal
main_container = widgets.VBox([
    widgets.HTML("<h2>Lector de Archivos de Texto (.txt)</h2>"),
    btn_abrir,
    widgets.HTML("<hr>"),
    text_area_content,
    output_area
], layout=widgets.Layout(align_items='center', border='1px solid #ccc', padding='20px', width='650px'))

# Mostrar la interfaz
display(main_container)
```

```python
# Instalaci√≥n silenciosa de Graphviz (si no est√°)
!apt-get install graphviz > /dev/null
!pip install graphviz > /dev/null

from graphviz import Digraph
from IPython.display import Image

# Crear el diagrama
dot = Digraph('CasosDeUso_Archivo', format='png')

# Configuraci√≥n general (vertical, estilo limpio)
dot.attr(rankdir='TB', bgcolor='#f8f9fa', splines='true', nodesep='1', ranksep='1')

# Actor tipo "stickman" con HTML
stickman_html = '''<
<TABLE BORDER="0" CELLBORDER="0" CELLPADDING="0">
  <TR><TD ALIGN="CENTER"><FONT FACE="monospace">O</FONT></TD></TR>
  <TR><TD ALIGN="CENTER"><FONT FACE="monospace">/|\\</FONT></TD></TR>
  <TR><TD ALIGN="CENTER"><FONT FACE="monospace">/ \\</FONT></TD></TR>
  <TR><TD ALIGN="CENTER">Usuario</TD></TR>
</TABLE>>'''
dot.node('actor', stickman_html, shape='none')

# Casos de uso (√≥valos)
dot.node('seleccionar', 'Seleccionar archivo .txt desde el sistema',
         shape='ellipse', style='filled', fillcolor='#cfe2f3', color='#6d9eeb')
dot.node('leer', 'Leer el contenido del archivo',
         shape='ellipse', style='filled', fillcolor='#b6d7a8', color='#38761d')
dot.node('mostrar', 'Mostrar el contenido en pantalla',
         shape='ellipse', style='filled', fillcolor='#f9cb9c', color='#e69138')
dot.node('error', 'Manejar error si el archivo no se puede leer',
         shape='ellipse', style='filled', fillcolor='#f4cccc', color='#990000')

# Conexiones
dot.edge('actor', 'seleccionar', color='#1c4587')
dot.edge('seleccionar', 'leer', color='#274e13')
dot.edge('leer', 'mostrar', color='#783f04')
dot.edge('leer', 'error', color='#990000')

# Renderizar y mostrar
output_path = dot.render('diagrama_casos_uso_archivo', format='png', cleanup=True)
Image(filename=output_path)
```
