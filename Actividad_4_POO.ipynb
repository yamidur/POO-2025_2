```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output

def ejecutar_prueba_excepciones():
    resultado = ""

    # Primer Bloque
    try:
        resultado += "Ingresando al primer try\n"
        cociente = 10000 / 0
        resultado += "Después de la división\n"
    except ZeroDivisionError:
        resultado += "División por cero\n"
    finally:
        resultado += "Ingresando al primer finally\n"

    # Segundo Bloque
    try:
        resultado += "Ingresando al segundo try\n"
        objeto = None
        objeto.toString()
        resultado += "Imprimiendo objeto\n"
    except ArithmeticError:
        resultado += "División por cero\n"
    except Exception:
        resultado += "Ocurrió una excepción\n"
    finally:
        resultado += "Ingresando al segundo finally\n"

    return resultado

# Se crea la ventana donde se muestran los resultados
output_area = widgets.Output()

# Elementos para el boton
def on_ejecutar_click(b):
    with output_area:
        # Limpia la salida
        clear_output(wait=True)
        print("--- Resultado de la Prueba de Excepciones ---")
        print(ejecutar_prueba_excepciones())
        print("---------------------------------------------")

# Crear los estilos
style = widgets.ButtonStyle(button_color='steelblue')

# Crea el layout para controlar el tamaño (ancho y alto)
layout = widgets.Layout(width='300px', height='50px', font_size='20px')

btn_ejecutar = widgets.Button(
    description="Ejecutar Prueba de Excepciones",
    style=style,
    layout=layout
)

# Asignar la acción al botón
btn_ejecutar.on_click(on_ejecutar_click)

# Mostrar la interfaz en la celda de Colab
display(btn_ejecutar, output_area)
```

Se requiere implementar una clase vendedor que posee los siguientes atri-
butos: nombre (tipo String), apellidos (tipo String) y edad (tipo int).

La clase contiene un constructor que inicialice los atributos de la clase. Además, la clase posee los siguientes métodos:

- Imprimir: muestra por pantalla los valores de sus atributos.

- Verificar edad: este método recibe como parámetro un valor entero que representa la edad del vendedor. Para que un vendedor pueda desempeñar sus labores se requiere que sea mayor de edad (mayor de 18 años). Si esta condición no se cumple, se lanza una excepción de tipo IllegalArgumentException con el mensaje “El vendedor debe ser mayor de 18 años”. Además, se evalúa si la edad

una excepción de tipo IllegalArgumentException con el mensaje “La
edad no puede ser negativa ni mayor a 120”. Si la edad cumple
estos requerimientos se pueden instanciar el objeto vendedor.

```python
# @title
import ipywidgets as widgets
from IPython.display import display, HTML
import re # Usado para validación simple de texto

# -----------------------------------------------------------------
# 1. Definición de la Lógica (Clase)
# -----------------------------------------------------------------

class VendedorAppLogic:
    """Clase que maneja la lógica de validación de los datos del vendedor."""
    def __init__(self, nombre, apellidos, edad):
        self.nombre = nombre
        self.apellidos = apellidos
        self.edad = edad

    def validar_y_obtener_info(self):
        """Valida la edad y devuelve la información formateada o levanta ValueError."""
        try:
            edad_valor = int(self.edad)
        except ValueError:
            raise ValueError("La edad debe ser un número entero.")

        if edad_valor < 18:
            raise ValueError("El vendedor debe ser mayor de 18 años.")

        # Asumiendo que el nombre y apellido no deben estar vacíos
        if not self.nombre.strip() or not self.apellidos.strip():
            raise ValueError("Nombre y Apellidos no pueden estar vacíos.")

        if not (0 <= edad_valor <= 120):
            raise ValueError("La edad debe estar entre 0 y 120 años.")

        return (
            f"Nombre: {self.nombre}\n"
            f"Apellidos: {self.apellidos}\n"
            f"Edad: {edad_valor} años"
        )

# -----------------------------------------------------------------
# 2. Interfaz Gráfica (Widgets)
# -----------------------------------------------------------------

# Definición de estilos
layout_ancho = widgets.Layout(width='200px')
estilo_label = widgets.Layout(width='100px', justify_content='flex-start')

# Widgets de entrada
titulo = widgets.HTML("<h2>Registro de Vendedor</h2>")

label_nombre = widgets.Label(value="Nombre:", layout=estilo_label)
input_nombre = widgets.Text(description="", layout=layout_ancho)
box_nombre = widgets.HBox([label_nombre, input_nombre])

label_apellidos = widgets.Label(value="Apellidos:", layout=estilo_label)
input_apellidos = widgets.Text(description="", layout=layout_ancho)
box_apellidos = widgets.HBox([label_apellidos, input_apellidos])

label_edad = widgets.Label(value="Edad:", layout=estilo_label)
input_edad = widgets.Text(description="", layout=layout_ancho) # Usamos Text para manejar la validación en el código
box_edad = widgets.HBox([label_edad, input_edad])

# Botón y Salida
boton_registrar = widgets.Button(description='Registrar', button_style='success', layout=widgets.Layout(width='150px'))
output_mensaje = widgets.Output()


# -----------------------------------------------------------------
# 3. Función de Manejo de Eventos (Callback)
# -----------------------------------------------------------------

def on_registrar_clicked(b):
    """Función que se ejecuta al hacer clic en el botón 'Registrar'."""
    with output_mensaje:
        output_mensaje.clear_output()

        # 1. Recuperar valores
        nombre = input_nombre.value
        apellidos = input_apellidos.value
        edad_str = input_edad.value

        try:
            # 2. Ejecutar la lógica de validación
            vendedor = VendedorAppLogic(nombre, apellidos, edad_str)
            info = vendedor.validar_y_obtener_info()

            # 3. Mostrar resultado exitoso (Simula messagebox.showinfo)
            print("--- REGISTRO EXITOSO ---")
            print(info)

        except ValueError as e:
            # 4. Mostrar error (Simula messagebox.showerror)
            display(HTML(f'<p style="color: red; font-weight: bold;">ERROR: {e}</p>'))

# Asociar el evento del botón a la función
boton_registrar.on_click(on_registrar_clicked)


# -----------------------------------------------------------------
# 4. Despliegue de la Interfaz
# -----------------------------------------------------------------

interfaz_completa = widgets.VBox([
    titulo,
    box_nombre,
    box_apellidos,
    box_edad,
    boton_registrar,
    widgets.HBox([widgets.Label(value="--- Resultados ---", layout=widgets.Layout(width='500px'))]),
    output_mensaje
], layout=widgets.Layout(border='1px solid lightgray', padding='20px', width='450px'))

display(interfaz_completa)
```

```python
# @title
import ipywidgets as widgets
from IPython.display import display, HTML
import math
import re # Usado para validación

# ============================================================
# CLASE DE LÓGICA (CalculosApp adaptada)
# ============================================================

class CalculadoraNumerica:
    """Clase para encapsular la lógica de cálculo de logaritmo y raíz cuadrada."""

    def __init__(self, valor):
        self.valor = valor

    def calcular_logaritmo(self):
        """Calcula el logaritmo natural."""
        return math.log(self.valor)

    def calcular_raiz_cuadrada(self):
        """Calcula la raíz cuadrada."""
        return math.sqrt(self.valor)

# ============================================================
# LÓGICA DE LA INTERFAZ (Reemplazando Tkinter con ipywidgets)
# ============================================================

# Widgets de Interfaz
output_area = widgets.Output()
valor_entry = widgets.Text(
    description='Ingrese un número positivo:',
    placeholder='Ej: 25.5',
    layout=widgets.Layout(width='300px')
)
btn_calcular = widgets.Button(
    description="Calcular",
    button_style='success',
    layout=widgets.Layout(width='120px')
)

# Área para mostrar resultados
resultado_ln = widgets.Label(value="")
resultado_sqrt = widgets.Label(value="")

def calcular_valores(btn):
    """Función llamada al presionar el botón 'Calcular'."""
    with output_area:
        output_area.clear_output()
        errores = []
        valor = None

        try:
            valor_str = valor_entry.value.strip()

            # 1. Validación de entrada numérica
            if not re.match(r'^-?\d+(\.\d+)?$', valor_str):
                 raise ValueError("El valor debe ser numérico.")

            valor = float(valor_str)

            # 2. Verificación de valor negativo (para los cálculos específicos)
            if valor < 0:
                errores.append("Error: El valor debe ser un número positivo para calcular el logaritmo.")
                errores.append("Error: El valor debe ser un número positivo para calcular la raíz cuadrada.")
                resultado_ln.value = "ln() = N/A"
                resultado_sqrt.value = "√() = N/A"
            else:
                # 3. Realizar cálculos
                calculadora = CalculadoraNumerica(valor)
                ln_val = calculadora.calcular_logaritmo()
                sqrt_val = calculadora.calcular_raiz_cuadrada()

                # Mostrar resultados
                resultado_ln.value = f"ln({valor}) = {ln_val:.4f}"
                resultado_sqrt.value = f"√{valor} = {sqrt_val:.4f}"

        except ValueError as e:
            # Capturar errores de conversión no numérica
            errores.append(f"Error: {e}")
            resultado_ln.value = "ln() = N/A"
            resultado_sqrt.value = "√() = N/A"

        except Exception as e:
            # Capturar otros errores (como log(0) o math.sqrt(negativo) si no se usa el if valor < 0)
            if "math domain error" in str(e):
                 errores.append("Error: El logaritmo de cero y la raíz de un negativo no están definidos en los reales.")
            else:
                errores.append(f"Error inesperado: {e}")
            resultado_ln.value = "ln() = N/A"
            resultado_sqrt.value = "√() = N/A"

        # Mostrar errores si los hay
        if errores:
            # Reemplazamos messagebox.showerror con una visualización HTML de error
            display(HTML(f'<div style="color: red; font-weight: bold; padding: 5px;">Errores encontrados:<br>- {"<br>- ".join(errores)}</div>'))
        elif valor is not None and valor >= 0:
             display(HTML(f'<div style="color: green; font-weight: bold;">✅ Cálculo completado.</div>'))


# Asignar la función al botón
btn_calcular.on_click(calcular_valores)

# Contenedor principal
main_container = widgets.VBox([
    widgets.HTML("<h2>Cálculos Numéricos (Logaritmo y Raíz Cuadrada)</h2>"),
    widgets.HBox([
        widgets.Label("Ingrese un número positivo:"),
        valor_entry
    ], layout=widgets.Layout(align_items='center', margin='10px 0')),

    btn_calcular,
    widgets.HTML("<hr style='width: 100%;'>"),

    widgets.VBox([
        widgets.HTML("<h3>Resultados:</h3>"),
        resultado_ln,
        resultado_sqrt
    ], layout=widgets.Layout(align_items='flex-start', border='1px solid #ddd', padding='15px', width='90%')),

    output_area
], layout=widgets.Layout(align_items='center', border='1px solid #ccc', padding='20px', width='500px'))

# Mostrar la interfaz
display(main_container)
```

```python
# @title

import ipywidgets as widgets
from IPython.display import display, HTML

# -----------------------------
# Clases
# -----------------------------
class Programador:
    def __init__(self, nombre, apellidos):
        self.nombre = nombre
        self.apellidos = apellidos

class EquipoMaratonProgramacion:
    def __init__(self, nombre_equipo, universidad, lenguaje_programacion):
        self.nombre_equipo = nombre_equipo
        self.universidad = universidad
        self.lenguaje_programacion = lenguaje_programacion
        self.programadores = []

    def anadir_programador(self, programador):
        if len(self.programadores) >= 3:
            raise Exception("El equipo no puede tener más de 3 programadores.")
        self.programadores.append(programador)

    @staticmethod
    def validar_campo(texto):
        if any(char.isdigit() for char in texto):
            raise Exception("El campo no puede contener dígitos.")
        if len(texto) > 20:
            raise Exception("El campo no debe tener más de 20 caracteres.")

# -----------------------------
# Widgets (interfaz Colab)
# -----------------------------
entry_nombre_equipo = widgets.Text(description="Equipo:", layout=widgets.Layout(width='400px'))
entry_universidad = widgets.Text(description="Universidad:", layout=widgets.Layout(width='400px'))
entry_lenguaje = widgets.Text(description="Lenguaje:", layout=widgets.Layout(width='400px'))

entries_nombres = []
entries_apellidos = []

# Crear campos para 3 integrantes
integrantes_widgets = []
for i in range(3):
    lbl_integrante = widgets.HTML(f"<h4>👨‍💻 Integrante {i+1}</h4>")
    entry_nombre = widgets.Text(description="Nombre:", layout=widgets.Layout(width='350px'))
    entry_apellido = widgets.Text(description="Apellidos:", layout=widgets.Layout(width='350px'))
    entries_nombres.append(entry_nombre)
    entries_apellidos.append(entry_apellido)
    integrantes_widgets.append(widgets.VBox([lbl_integrante, entry_nombre, entry_apellido]))

# Botón y salida
btn_registrar = widgets.Button(description="Registrar equipo", button_style='success')
output = widgets.Output()

# -----------------------------
# Lógica de registro
# -----------------------------
def registrar_equipo(b):
    with output:
        output.clear_output()
        try:
            nombre_equipo = entry_nombre_equipo.value.strip()
            universidad = entry_universidad.value.strip()
            lenguaje = entry_lenguaje.value.strip()

            if not nombre_equipo or not universidad or not lenguaje:
                raise Exception("Todos los campos del equipo son obligatorios.")

            equipo = EquipoMaratonProgramacion(nombre_equipo, universidad, lenguaje)

            programadores_validos = 0
            for i in range(3):
                nombre = entries_nombres[i].value.strip()
                apellidos = entries_apellidos[i].value.strip()

                if nombre and apellidos:
                    EquipoMaratonProgramacion.validar_campo(nombre)
                    EquipoMaratonProgramacion.validar_campo(apellidos)
                    equipo.anadir_programador(Programador(nombre, apellidos))
                    programadores_validos += 1
                elif nombre or apellidos:
                    raise Exception(f"Debes completar nombre y apellidos del integrante {i+1} o dejar ambos en blanco.")

            if programadores_validos < 2:
                raise Exception("Debes registrar al menos 2 programadores.")

            print(f"✅ Equipo '{equipo.nombre_equipo}' registrado correctamente con {programadores_validos} integrantes.")
        except Exception as e:
            print(f"❌ Error: {e}")

btn_registrar.on_click(registrar_equipo)

# -----------------------------
# Mostrar interfaz
# -----------------------------
display(HTML("<h3 style='color:#1a5276'>Ejercicio 6.7 - Registro de Equipo (mínimo 2 integrantes)</h3>"))
display(HTML("<b>Datos del Equipo</b>"))
display(entry_nombre_equipo, entry_universidad, entry_lenguaje)
display(HTML("<hr><b>Datos de los Programadores (mínimo 2)</b>"))

for box in integrantes_widgets:
    display(box)

display(btn_registrar, output)
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML
import io
from google.colab import files # Necesario para la subida de archivos en Colab

# ============================================================
# LECTOR DE ARCHIVOS PARA COLAB
# ============================================================

# Widgets de Interfaz
output_area = widgets.Output()
text_area_content = widgets.Textarea(
    value='',
    placeholder='El contenido del archivo se mostrará aquí...',
    description='Contenido:',
    disabled=False,
    layout=widgets.Layout(width='95%', height='300px')
)
btn_abrir = widgets.Button(
    description="Abrir archivo .txt",
    button_style='primary',
    tooltip='Sube un archivo .txt',
    layout=widgets.Layout(width='200px')
)

def leer_archivo_colab(btn):
    """
    Función que simula filedialog.askopenfilename de Tkinter usando
    el módulo 'files' de Google Colab para la subida.
    """
    text_area_content.value = 'Cargando archivo...'
    with output_area:
        output_area.clear_output()

        try:
            # Sube un archivo. El usuario seleccionará el archivo en un diálogo
            uploaded = files.upload()
        except Exception:
            # Esto ocurre si el usuario cancela la selección
            text_area_content.value = 'Selección de archivo cancelada.'
            return

        # Procesar el archivo subido
        if not uploaded:
            text_area_content.value = 'Selección de archivo cancelada.'
            return

        # Tomar el nombre del primer archivo subido
        nombre_archivo = list(uploaded.keys())[0]

        # Verificar si es un archivo .txt
        if not nombre_archivo.lower().endswith('.txt'):
            display(HTML('<div style="color: red; font-weight: bold;">❌ Error: Por favor, seleccione un archivo con extensión .txt.</div>'))
            text_area_content.value = 'Error: Archivo no compatible.'
            return

        try:
            # Leer el contenido del archivo subido
            contenido = uploaded[nombre_archivo].decode('utf-8')

            # Mostrar el contenido en el Textarea
            text_area_content.value = contenido
            display(HTML(f'<div style="color: green;">✅ Archivo "{nombre_archivo}" cargado exitosamente.</div>'))

        except Exception as e:
            display(HTML(f'<div style="color: red; font-weight: bold;">❌ Error de lectura: No se pudo leer el archivo. ({e})</div>'))
            text_area_content.value = f"Error al procesar el archivo: {e}"

# Asignar la función al botón
btn_abrir.on_click(leer_archivo_colab)

# Contenedor principal
main_container = widgets.VBox([
    widgets.HTML("<h2>Lector de Archivos de Texto (.txt)</h2>"),
    btn_abrir,
    widgets.HTML("<hr>"),
    text_area_content,
    output_area
], layout=widgets.Layout(align_items='center', border='1px solid #ccc', padding='20px', width='650px'))

# Mostrar la interfaz
display(main_container)
```
