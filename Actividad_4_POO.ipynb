```python
import ipywidgets as widgets
from IPython.display import display, clear_output

class PruebaExcepcionesApp:
    """
    Clase que encapsula la lógica de la prueba de excepciones
    """

    def __init__(self):
        # 1. Componentes de la Interfaz
        self.output_area = widgets.Output()
        self.btn_ejecutar = self._crear_boton_ejecutar()

        # 2. Conexión de Eventos
        self.btn_ejecutar.on_click(self._on_ejecutar_click)

    def _crear_boton_ejecutar(self):
        style = widgets.ButtonStyle(button_color='steelblue')
        layout = widgets.Layout(width='300px', height='50px', font_size='20px')

        btn = widgets.Button(
            description="Ejecutar Prueba de Excepciones (POO)",
            style=style,
            layout=layout
        )
        return btn

    def _ejecutar_prueba_excepciones(self):
        resultado = ""

        # Primer Bloque
        try:
            resultado += "Ingresando al primer try\n"
            cociente = 10000 / 0
            resultado += "Después de la división\n"
        except ZeroDivisionError:
            resultado += "División por cero\n"
        finally:
            resultado += "Ingresando al primer finally\n"

        # Segundo Bloque
        try:
            resultado += "Ingresando al segundo try\n"
            objeto = None
            objeto.capitalize()
            resultado += "Imprimiendo objeto\n"
        except ArithmeticError:
            resultado += "División por cero\n"
        except Exception:
            resultado += "Ocurrió una excepción\n"
        finally:
            resultado += "Ingresando al segundo finally\n"

        return resultado

    def _on_ejecutar_click(self, b):
        """
        Manejador de eventos para el clic del botón.
        """
        # Accedemos al área de salida de la instancia usando self.output_area
        with self.output_area:
            clear_output(wait=True)
            print("--- Resultado de la Prueba de Excepciones (POO) ---")
            print(self._ejecutar_prueba_excepciones())
            print("-----------------------------------------------------")

    def mostrar_interfaz(self):
        """
        Muestra los widgets creados por la instancia.
        """
        display(self.btn_ejecutar, self.output_area)

# Instanciación y Ejecución de la Aplicación
app = PruebaExcepcionesApp()

# Se invoca el método para mostrar la interfaz
app.mostrar_interfaz()
```

Se requiere implementar una clase vendedor que posee los siguientes atri-
butos: nombre (tipo String), apellidos (tipo String) y edad (tipo int).

La clase contiene un constructor que inicialice los atributos de la clase. Además, la clase posee los siguientes métodos:

- Imprimir: muestra por pantalla los valores de sus atributos.

- Verificar edad: este método recibe como parámetro un valor entero que representa la edad del vendedor. Para que un vendedor pueda desempeñar sus labores se requiere que sea mayor de edad (mayor de 18 años). Si esta condición no se cumple, se lanza una excepción de tipo IllegalArgumentException con el mensaje “El vendedor debe ser mayor de 18 años”. Además, se evalúa si la edad
se encuentra en el rango de 0 a 120, si no se cumple, se genera
una excepción de tipo IllegalArgumentException con el mensaje “La
edad no puede ser negativa ni mayor a 120”. Si la edad cumple
estos requerimientos se pueden instanciar el objeto vendedor.
Además, se requiere que los datos del vendedor se ingresen por
teclado.

```python
import ipywidgets as widgets
from IPython.display import display, HTML
import re # Usado para validación simple de texto

# Definición de la clase

class VendedorAppLogic:
    """Clase que maneja la lógica de validación de los datos del vendedor."""
    def __init__(self, nombre, apellidos, edad):
        self.nombre = nombre
        self.apellidos = apellidos
        self.edad = edad

    def validar_y_obtener_info(self):
        """Valida la edad y devuelve la información formateada o levanta ValueError."""
        try:
            edad_valor = int(self.edad)
        except ValueError:
            raise ValueError("La edad debe ser un número entero.")

        if edad_valor < 18:
            raise ValueError("El vendedor debe ser mayor de 18 años.")

        # Asumiendo que el nombre y apellido no deben estar vacíos
        if not self.nombre.strip() or not self.apellidos.strip():
            raise ValueError("Nombre y Apellidos no pueden estar vacíos.")

        if not (0 <= edad_valor <= 120):
            raise ValueError("La edad debe estar entre 0 y 120 años.")

        return (
            f"Nombre: {self.nombre}\n"
            f"Apellidos: {self.apellidos}\n"
            f"Edad: {edad_valor} años"
        )

# Interfaz Gráfica (Widgets)

# Definición de estilos
layout_ancho = widgets.Layout(width='200px')
estilo_label = widgets.Layout(width='100px', justify_content='flex-start')

# Widgets de entrada
titulo = widgets.HTML("<h2>Registro de Vendedor</h2>")

label_nombre = widgets.Label(value="Nombre:", layout=estilo_label)
input_nombre = widgets.Text(description="", layout=layout_ancho)
box_nombre = widgets.HBox([label_nombre, input_nombre])

label_apellidos = widgets.Label(value="Apellidos:", layout=estilo_label)
input_apellidos = widgets.Text(description="", layout=layout_ancho)
box_apellidos = widgets.HBox([label_apellidos, input_apellidos])

label_edad = widgets.Label(value="Edad:", layout=estilo_label)
input_edad = widgets.Text(description="", layout=layout_ancho)
box_edad = widgets.HBox([label_edad, input_edad])

# Botón y Salida
boton_registrar = widgets.Button(description='Registrar', button_style='success', layout=widgets.Layout(width='150px'))
output_mensaje = widgets.Output()

# 3. Función de Manejo de Eventos (Callback)

def on_registrar_clicked(b):
    """Función que se ejecuta al hacer clic en el botón 'Registrar'."""
    with output_mensaje:
        output_mensaje.clear_output()

        nombre = input_nombre.value
        apellidos = input_apellidos.value
        edad_str = input_edad.value

        try:
            # Ejecutar la validación
            vendedor = VendedorAppLogic(nombre, apellidos, edad_str)
            info = vendedor.validar_y_obtener_info()

            # Mostrar resultado exitoso
            print("--- REGISTRO EXITOSO ---")
            print(info)

        except ValueError as e:
            # Mostrar error
            display(HTML(f'<p style="color: red; font-weight: bold;">ERROR: {e}</p>'))

# Asociar el evento del botón a la función
boton_registrar.on_click(on_registrar_clicked)

# 4. Despliegue de la Interfaz

interfaz_completa = widgets.VBox([
    titulo,
    box_nombre,
    box_apellidos,
    box_edad,
    boton_registrar,
    widgets.HBox([widgets.Label(value="--- Resultados ---", layout=widgets.Layout(width='500px'))]),
    output_mensaje
], layout=widgets.Layout(border='1px solid lightgray', padding='20px', width='450px'))

display(interfaz_completa)
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import math
import re

# Definicion de la clase

class CalculadoraNumerica:
    """Clase para encapsular la lógica pura de cálculo (Modelo)."""

    def __init__(self, valor):
        self.valor = valor

    def calcular_logaritmo(self):
        """Calcula el logaritmo natural."""
        return math.log(self.valor)

    def calcular_raiz_cuadrada(self):
        """Calcula la raíz cuadrada."""
        return math.sqrt(self.valor)

# CLASE DE APLICACIÓN (Vista y Controlador)

class CalculadoraApp:
    """
    Clase que encapsula la interfaz de usuario y manejo de errores.
    """

    def __init__(self):
        self.output_area = widgets.Output()
        self.valor_entry = widgets.Text(
            placeholder='Ej: 25.5',
            layout=widgets.Layout(width='200px')
        )
        self.btn_calcular = widgets.Button(
            description="Calcular",
            button_style='success',
            layout=widgets.Layout(width='120px')
        )
        self.resultado_ln = widgets.Label(value="")
        self.resultado_sqrt = widgets.Label(value="")

        self.btn_calcular.on_click(self._on_calcular_click)

        self.main_container = self._crear_interfaz()

    def _crear_interfaz(self):
        """Construye y devuelve el contenedor principal de la interfaz."""
        input_row = widgets.HBox([
            widgets.Label("Ingrese un número positivo:"),
            self.valor_entry
        ], layout=widgets.Layout(align_items='center', margin='10px 0'))

        results_box = widgets.VBox([
            widgets.HTML("<h3>Resultados:</h3>"),
            self.resultado_ln,
            self.resultado_sqrt
        ], layout=widgets.Layout(align_items='flex-start', border='1px solid #ddd', padding='15px', width='90%'))

        return widgets.VBox([
            widgets.HTML("<h2>Cálculos Numéricos (POO)</h2>"),
            input_row,
            self.btn_calcular,
            widgets.HTML("<hr style='width: 100%;'>"),
            results_box,
            self.output_area
        ], layout=widgets.Layout(align_items='center', border='1px solid #ccc', padding='20px', width='550px'))

    def _mostrar_error(self, errores):
        """Muestra los errores formateados en el output_area."""
        html_error = f'<div style="color: red; font-weight: bold; padding: 5px;">Errores encontrados:<br>- {"<br>- ".join(errores)}</div>'
        display(HTML(html_error))

    def _mostrar_exito(self):
        """Muestra un mensaje de éxito."""
        display(HTML(f'<div style="color: green; font-weight: bold;">✅ Cálculo completado.</div>'))

    def _on_calcular_click(self, btn):
        """Manejador de eventos que encapsula toda la lógica de control."""
        clear_output(wait=True)
        errores = []
        valor = None

        try:
            valor_str = self.valor_entry.value.strip()

            if not valor_str:
                 raise ValueError("Debe ingresar un valor.")

            if not re.match(r'^-?\d+(\.\d+)?$', valor_str):
                raise ValueError("El valor debe ser un número válido.")

            valor = float(valor_str)

            calculadora = CalculadoraNumerica(valor)

            self.resultado_ln.value = "ln() = N/A"
            self.resultado_sqrt.value = "√() = N/A"

            # Logaritmo
            try:
                ln_val = calculadora.calcular_logaritmo()
                self.resultado_ln.value = f"ln({valor}) = {ln_val:.4f}"
            except ValueError:
                errores.append("Error ln: El logaritmo de un número negativo o cero no está definido.")

            # Raíz Cuadrada
            try:
                sqrt_val = calculadora.calcular_raiz_cuadrada()
                self.resultado_sqrt.value = f"√{valor} = {sqrt_val:.4f}"
            except ValueError:
                errores.append("Error √: La raíz cuadrada de un número negativo no está definida.")

        except ValueError as e:
            errores.append(f"Error de entrada: {e}")
            self.resultado_ln.value = "ln() = N/A"
            self.resultado_sqrt.value = "√() = N/A"

        except Exception as e:
            errores.append(f"Error inesperado: {e}")
            self.resultado_ln.value = "ln() = N/A"
            self.resultado_sqrt.value = "√() = N/A"

        with self.output_area:
            clear_output(wait=True)
            if errores:
                self._mostrar_error(errores)
            elif valor is not None:
                self._mostrar_exito()

    def display_app(self):
        """Método para mostrar la interfaz en un entorno Jupyter/Colab."""
        display(self.main_container)

# Instanciación y Ejecución
if __name__ == '__main__':
    app = CalculadoraApp()
    app.display_app()
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

# Definicion de clases

class Programador:
    """Representa a un miembro del equipo."""
    def __init__(self, nombre, apellidos):
        self.nombre = nombre
        self.apellidos = apellidos

class EquipoMaratonProgramacion:
    """Representa al equipo de programación con su lógica de validación."""
    def __init__(self, nombre_equipo, universidad, lenguaje_programacion):
        self.nombre_equipo = nombre_equipo
        self.universidad = universidad
        self.lenguaje_programacion = lenguaje_programacion
        self.programadores = []

    def anadir_programador(self, programador):
        """Añade un programador al equipo, verificando el límite."""
        if len(self.programadores) >= 3:
            raise ValueError("El equipo no puede tener más de 3 programadores.")
        self.programadores.append(programador)

    @staticmethod
    def validar_campo(texto, campo_nombre="campo"):
        """Método estático para validar campos de texto."""
        if not texto:
            raise ValueError(f"El campo '{campo_nombre}' es obligatorio.")
        if any(char.isdigit() for char in texto):
            raise ValueError(f"El campo '{campo_nombre}' no puede contener dígitos.")
        if len(texto) > 20:
            raise ValueError(f"El campo '{campo_nombre}' no debe tener más de 20 caracteres.")

# CLASE DE LA APLICACIÓN (Vista y Controlador)

class RegistroEquipoApp:
    """
    Clase que encapsula la interfaz de usuario (Widgets) y la lógica
    de control (Manejo de eventos y validación).
    """

    MAX_PROGRAMADORES = 3
    MIN_PROGRAMADORES = 2

    def __init__(self):
        self.output = widgets.Output()
        self.entry_nombre_equipo = widgets.Text(description="Equipo:", layout=widgets.Layout(width='400px'))
        self.entry_universidad = widgets.Text(description="Universidad:", layout=widgets.Layout(width='400px'))
        self.entry_lenguaje = widgets.Text(description="Lenguaje:", layout=widgets.Layout(width='400px'))
        self.entries_nombres = []
        self.entries_apellidos = []

        self.btn_registrar = widgets.Button(description="Registrar equipo", button_style='success')
        self.btn_registrar.on_click(self._on_registrar_click)

        self.main_container = self._crear_interfaz()


    def _crear_integrante_widgets(self, i):
        """Crea y almacena los widgets para un solo integrante."""
        lbl_integrante = widgets.HTML(f"<h4>👨‍💻 Integrante {i+1}</h4>")
        entry_nombre = widgets.Text(description="Nombre:", layout=widgets.Layout(width='350px'))
        entry_apellido = widgets.Text(description="Apellidos:", layout=widgets.Layout(width='350px'))

        self.entries_nombres.append(entry_nombre)
        self.entries_apellidos.append(entry_apellido)

        return widgets.VBox([lbl_integrante, entry_nombre, entry_apellido])

    def _crear_interfaz(self):
        """Construye y devuelve el contenedor principal."""
        integrantes_widgets = [self._crear_integrante_widgets(i) for i in range(self.MAX_PROGRAMADORES)]

        equipo_box = widgets.VBox([
            widgets.HTML("<b>Datos del Equipo</b>"),
            self.entry_nombre_equipo,
            self.entry_universidad,
            self.entry_lenguaje
        ])

        programadores_box = widgets.VBox([
            widgets.HTML("<hr><b>Datos de los Programadores (mínimo 2)</b>"),
            *integrantes_widgets
        ])

        return widgets.VBox([
            widgets.HTML("<h3 style='color:#1a5276'>Ejercicio 6.7 - Registro de Equipo (POO)</h3>"),
            equipo_box,
            programadores_box,
            self.btn_registrar,
            self.output
        ])

    def _on_registrar_click(self, b):
        """Manejador de eventos: procesa la entrada y registra el equipo."""
        with self.output:
            clear_output()
            try:
                nombre_equipo = self.entry_nombre_equipo.value.strip()
                universidad = self.entry_universidad.value.strip()
                lenguaje = self.entry_lenguaje.value.strip()

                EquipoMaratonProgramacion.validar_campo(nombre_equipo, "nombre de equipo")
                EquipoMaratonProgramacion.validar_campo(universidad, "universidad")
                if not lenguaje:
                    raise ValueError("El campo Lenguaje es obligatorio.")

                equipo = EquipoMaratonProgramacion(nombre_equipo, universidad, lenguaje)

                programadores_validos = 0
                for i in range(self.MAX_PROGRAMADORES):
                    nombre = self.entries_nombres[i].value.strip()
                    apellidos = self.entries_apellidos[i].value.strip()

                    if nombre and apellidos:
                        EquipoMaratonProgramacion.validar_campo(nombre, f"nombre de integrante {i+1}")
                        EquipoMaratonProgramacion.validar_campo(apellidos, f"apellidos de integrante {i+1}")

                        equipo.anadir_programador(Programador(nombre, apellidos))
                        programadores_validos += 1
                    elif nombre or apellidos:
                        raise ValueError(f"Debes completar el nombre y apellidos del integrante {i+1} o dejar AMBOS en blanco.")

                if programadores_validos < self.MIN_PROGRAMADORES:
                    raise ValueError(f"Debes registrar al menos {self.MIN_PROGRAMADORES} programadores.")

                # Éxito
                display(HTML(f'<p style="color: green; font-weight: bold;">✅ Equipo "{equipo.nombre_equipo}" registrado con {programadores_validos} integrante(s).</p>'))

            except ValueError as e:
                display(HTML(f'<p style="color: red; font-weight: bold;"> Error de Validación: {e}</p>'))

            except Exception as e:
                display(HTML(f'<p style="color: red; font-weight: bold;"> Error: {e}</p>'))

    def display_app(self):
        """Método para mostrar la interfaz completa."""
        display(self.main_container)

# Ejecucion
if __name__ == '__main__':
    app = RegistroEquipoApp()
    app.display_app()
```

```python
# @title
# ============================================================
# DIAGRAMA UML: CADENA DE LECTURA DE ARCHIVOS (FileInputStream)
# ============================================================

# Instalación silenciosa de Graphviz
!apt-get install -y graphviz > /dev/null 2>&1
!pip install graphviz > /dev/null 2>&1

from graphviz import Digraph
from IPython.display import Image, display

# Crear el diagrama
uml = Digraph('CadenaLecturaArchivos', filename='uml_lectura_archivos', format='png')

# --------------------------------------------------------------
# Configuración general y estilo
# --------------------------------------------------------------
uml.attr(
    rankdir='LR',
    ranksep='0.8',
    nodesep='0.5',
    bgcolor='#f8f9fa' # Fondo claro
)
uml.attr('graph', dpi='100')

# Estilo base para los nodos de clase
uml.attr('node',
          shape='record',
          style='filled',
          color='#000000',
          fontname='Helvetica',
          fontsize='11')

# --------------------------------------------------------------
# 1. Clase Principal (LeerArchivo)
# --------------------------------------------------------------
# Contenido de la clase: Nombre y el método principal
lector_content = '''{
    LeerArchivo
    |
    +main(String[] args)
}'''
uml.node('Lector',
         lector_content,
         fillcolor='#CCFFCC', # Verde pálido para la clase de aplicación
         margin='0.15')

# --------------------------------------------------------------
# 2. Clases de Stream/Buffer
# --------------------------------------------------------------
# Color de relleno diferente para las clases de librería
stream_color = '#CCE5FF' # Azul pálido

# FileInputStream
uml.node('FileInputStream',
         '{FileInputStream |}',
         fillcolor=stream_color)

# InputStreamReader
uml.node('InputStreamReader',
         '{InputStreamReader |}',
         fillcolor=stream_color)

# BufferedReader
uml.node('BufferedReader',
         '{BufferedReader |}',
         fillcolor=stream_color)

# --------------------------------------------------------------
# 3. Relaciones (Asociaciones)
# --------------------------------------------------------------
# Estilo de la flecha: Asociación (flecha abierta) con línea punteada
uml.attr('edge', arrowhead='open', style='dashed', color='#555555')

# LeerArchivo -> FileInputStream
uml.edge('Lector', 'FileInputStream')

# FileInputStream -> InputStreamReader
uml.edge('FileInputStream', 'InputStreamReader')

# InputStreamReader -> BufferedReader
uml.edge('InputStreamReader', 'BufferedReader')

# --------------------------------------------------------------
# Generar y mostrar el diagrama
# --------------------------------------------------------------
output_path = uml.render(cleanup=True)
display(Image(filename=output_path))
```

```python
# @title
import ipywidgets as widgets
from IPython.display import display, HTML
import io

# CLASE LECTOR DE ARCHIVOS

class LectorArchivos:

    def __init__(self):
        # Interfaz
        self.output_area = widgets.Output()
        self.text_area_content = widgets.Textarea(
            value='',
            placeholder='El contenido del archivo se mostrará aquí...',
            description='Contenido:',
            disabled=False,
            layout=widgets.Layout(width='95%', height='300px')
        )
        self.btn_abrir = widgets.Button(
            description="Abrir archivo .txt",
            button_style='primary',
            tooltip='Sube un archivo .txt',
            layout=widgets.Layout(width='200px')
        )

        # Asignacion del boton
        self.btn_abrir.on_click(self.leer_archivo_colab)

        # Contenedor principal
        self.main_container = widgets.VBox([
            widgets.HTML("<h2>Lector de Archivos de Texto (.txt)</h2>"),
            self.btn_abrir,
            widgets.HTML("<hr>"),
            self.text_area_content,
            self.output_area
        ], layout=widgets.Layout(align_items='center', border='1px solid #ccc', padding='20px', width='650px'))

    def mostrar_mensaje(self, mensaje, tipo='success'):
        with self.output_area:
            self.output_area.clear_output()
            style = ""
            if tipo == 'success':
                style = "color: green;"
                prefijo = "Correcto"
            elif tipo == 'error':
                style = "color: red; font-weight: bold;"
                prefijo = "Error:"

            display(HTML(f'<div style="{style}">{prefijo} {mensaje}</div>'))

    def leer_archivo_colab(self, btn):
        self.text_area_content.value = 'Cargando archivo...'

        try:
            # Sube un archivo.
            uploaded = files.upload()
        except Exception:
            # Ocurre si el usuario cancela la selección
            self.text_area_content.value = 'Selección de archivo cancelada.'
            self.mostrar_mensaje("Selección de archivo cancelada.", tipo='error')
            return

        if not uploaded:
            self.text_area_content.value = 'Selección de archivo cancelada.'
            self.mostrar_mensaje("Selección de archivo cancelada.", tipo='error')
            return

        nombre_archivo = list(uploaded.keys())[0]

        # Validación de extensión
        if not nombre_archivo.lower().endswith('.txt'):
            self.mostrar_mensaje(f"Por favor, seleccione un archivo con extensión .txt. (Se seleccionó {nombre_archivo})", tipo='error')
            self.text_area_content.value = 'Error: Archivo no compatible.'
            return

        # Lectura y visualización
        try:
            # Leer el contenido del archivo subido
            contenido = uploaded[nombre_archivo].decode('utf-8')

            # Mostrar el contenido
            self.text_area_content.value = contenido
            self.mostrar_mensaje(f'Archivo "{nombre_archivo}" cargado exitosamente.', tipo='success')

        except Exception as e:
            self.mostrar_mensaje(f'No se pudo leer el archivo. Detalle: ({e})', tipo='error')
            self.text_area_content.value = f"Error al procesar el archivo: {e}"

    def display_ui(self):
        """Muestra la interfaz de usuario en el entorno de notebook."""
        display(self.main_container)

# EJECUCIÓN

lector = LectorArchivos()
lector.display_ui()
```

```python
# @title
# Instalación silenciosa de Graphviz (si no está)
!apt-get install graphviz > /dev/null
!pip install graphviz > /dev/null

from graphviz import Digraph
from IPython.display import Image

# Crear el diagrama
dot = Digraph('CasosDeUso_Archivo', format='png')

# Configuración general (vertical, estilo limpio)
dot.attr(rankdir='LR', bgcolor='#f8f9fa', splines='true', nodesep='1', ranksep='1')

# Actor tipo "stickman" con HTML
stickman_html = '''<
<TABLE BORDER="0" CELLBORDER="0" CELLPADDING="0">
  <TR><TD ALIGN="CENTER"><FONT FACE="monospace">O</FONT></TD></TR>
  <TR><TD ALIGN="CENTER"><FONT FACE="monospace">/|\\</FONT></TD></TR>
  <TR><TD ALIGN="CENTER"><FONT FACE="monospace">/ \\</FONT></TD></TR>
  <TR><TD ALIGN="CENTER">Usuario</TD></TR>
</TABLE>>'''
dot.node('actor', stickman_html, shape='none')

dot.attr('graph', dpi='60')

# Casos de uso (óvalos)
dot.node('seleccionar', 'Seleccionar archivo .txt desde el sistema',
         shape='ellipse', style='filled', fillcolor='#cfe2f3', color='#6d9eeb')
dot.node('leer', 'Leer el contenido del archivo',
         shape='ellipse', style='filled', fillcolor='#b6d7a8', color='#38761d')
dot.node('mostrar', 'Mostrar el contenido en pantalla',
         shape='ellipse', style='filled', fillcolor='#f9cb9c', color='#e69138')
dot.node('error', 'Manejar error si el archivo no se puede leer',
         shape='ellipse', style='filled', fillcolor='#f4cccc', color='#990000')

# Conexiones
dot.edge('actor', 'seleccionar', color='#1c4587')
dot.edge('seleccionar', 'leer', color='#274e13')
dot.edge('leer', 'mostrar', color='#783f04')
dot.edge('leer', 'error', color='#990000')

# Renderizar y mostrar
output_path = dot.render('diagrama_casos_uso_archivo', format='png', cleanup=True)
Image(filename=output_path)
```
