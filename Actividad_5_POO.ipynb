# Actividad 5
Aplicación de Interfaz Grafica
(formulario con botones create, read, update y delete)

```python
# @title
import ipywidgets as widgets
from IPython.display import display, HTML
import os

# --- Configuración y Nombre del Archivo ---
FILE_NAME = "Friends_Contact.txt"

# --- Funciones de Utilidad para el Manejo de Archivos (Sin cambios) ---

def get_all_records():
    """Lee todos los registros del archivo y retorna una lista de tuplas (nombre, telefono)."""
    if not os.path.exists(FILE_NAME):
        with open(FILE_NAME, 'w') as f:
            pass
        return []

    records = []
    try:
        with open(FILE_NAME, 'r') as f:
            lines = [line.strip() for line in f if line.strip()]
            for line in lines:
                try:
                    nombre, telefono = line.split(',', 1)
                    records.append((nombre, telefono))
                except ValueError:
                    continue
    except Exception as e:
        print(f"ERROR: No se pudo leer el archivo. {e}")
        return []

    return records

def rewrite_file(records):
    """Escribe una lista de tuplas (nombre, telefono) de nuevo al archivo."""
    try:
        with open(FILE_NAME, 'w') as f:
            for nombre, telefono in records:
                f.write(f"{nombre},{telefono}\n")
    except Exception as e:
        print(f"ERROR: No se pudo reescribir el archivo. {e}")

# --- Componentes de la Interfaz (ipywidgets) ---

# Etiquetas
label_nombre = widgets.Label("Nombre:")
label_telefono = widgets.Label("Teléfono:")

# Campos de entrada
entry_nombre = widgets.Text(value='', placeholder='Nombre del Contacto', description_tooltip='Nombre')
entry_telefono = widgets.Text(value='', placeholder='Número de Teléfono', description_tooltip='Teléfono')

# Botón de Salida para mensajes
output_log = widgets.Output()

# --- Definición de Estilos de Botones con Colores ---

# Estilos CRUD (Colores de acción)
style_create = widgets.ButtonStyle(button_color='#4CAF50') # Verde
style_read = widgets.ButtonStyle(button_color='#4CAF50')   # Azul
style_update = widgets.ButtonStyle(button_color='#4CAF50') # Amarillo/Naranja
style_delete = widgets.ButtonStyle(button_color='#4CAF50') # Rojo
style_clean = widgets.ButtonStyle(button_color='#A9A9A9')  # Gris oscuro para Limpiar

# Botones con los nuevos estilos
btn_create = widgets.Button(description="Crear", style=style_create)
btn_read = widgets.Button(description="Leer", style=style_read)
btn_update = widgets.Button(description="Actualizar", style=style_update)
btn_delete = widgets.Button(description="Eliminar", style=style_delete)
btn_clean = widgets.Button(description="Limpiar Campos", style=style_clean)


# --- Lógica de Manejo de Eventos (Sin cambios en la lógica) ---

def clean_fields(b):
    """Limpia los campos y el log."""
    entry_nombre.value = ''
    entry_telefono.value = ''
    with output_log:
        output_log.clear_output()
        print(">> Campos limpiados.")

def create_record_colab(b):
    nombre = entry_nombre.value.strip()
    telefono = entry_telefono.value.strip()
    with output_log:
        output_log.clear_output()
        if not nombre or not telefono:
            print("ERROR: Ambos campos son requeridos.")
            return

        records = get_all_records()
        if any(n == nombre for n, t in records):
            print(f"ADVERTENCIA: '{nombre}' ya existe. Intente 'Actualizar'.")
            return

        records.append((nombre, telefono))
        rewrite_file(records)
        print(f"ÉXITO: Contacto '{nombre}' creado.")
        clean_fields(None)

def read_record_colab(b):
    nombre_to_find = entry_nombre.value.strip()
    with output_log:
        output_log.clear_output()
        if not nombre_to_find:
            print("ADVERTENCIA: Ingrese el Nombre a buscar.")
            return

        records = get_all_records()
        found_record = next(((n, t) for n, t in records if n == nombre_to_find), None)

        if found_record:
            entry_nombre.value = found_record[0]
            entry_telefono.value = found_record[1]
            print(f"ÉXITO: Contacto '{nombre_to_find}' cargado.")
        else:
            print(f"ERROR: Contacto '{nombre_to_find}' no encontrado.")
            entry_telefono.value = ''

def update_record_colab(b):
    nombre_to_update = entry_nombre.value.strip()
    new_telefono = entry_telefono.value.strip()
    with output_log:
        output_log.clear_output()
        if not nombre_to_update or not new_telefono:
            print("ERROR: Ambos campos son requeridos para actualizar.")
            return

        records = get_all_records()
        updated = False
        new_records = []

        for nombre, telefono in records:
            if nombre == nombre_to_update:
                new_records.append((nombre, new_telefono))
                updated = True
            else:
                new_records.append((nombre, telefono))

        if updated:
            rewrite_file(new_records)
            print(f"ÉXITO: Contacto '{nombre_to_update}' actualizado.")
            clean_fields(None)
        else:
            print(f"ERROR: Contacto '{nombre_to_update}' no encontrado para actualizar.")

def delete_record_colab(b):
    nombre_to_delete = entry_nombre.value.strip()
    with output_log:
        output_log.clear_output()
        if not nombre_to_delete:
            print("ADVERTENCIA: Ingrese el Nombre a eliminar.")
            return

        records = get_all_records()
        initial_count = len(records)
        new_records = [(n, t) for n, t in records if n != nombre_to_delete]

        if len(new_records) < initial_count:
            rewrite_file(new_records)
            print(f"ÉXITO: Contacto '{nombre_to_delete}' eliminado.")
            clean_fields(None)
        else:
            print(f"ERROR: Contacto '{nombre_to_delete}' no encontrado para eliminar.")

# --- Conectar las Funciones a los Botones ---
btn_create.on_click(create_record_colab)
btn_read.on_click(read_record_colab)
btn_update.on_click(update_record_colab)
btn_delete.on_click(delete_record_colab)
btn_clean.on_click(clean_fields)

# --- Estructura Visual ---

# Diseño de las entradas (Nombre y Teléfono)
input_layout = widgets.VBox([
    widgets.HBox([label_nombre, entry_nombre]),
    widgets.HBox([label_telefono, entry_telefono])
])

# Diseño de los botones CRUD (similar a una grilla 2x2)
crud_buttons = widgets.GridBox(children=[
    btn_update, btn_read,
    btn_create, btn_delete
], layout=widgets.Layout(grid_template_columns="1fr 1fr", width='500px'))

# Layout final con color de fondo gris claro (#EEEEEE)
app_layout = widgets.VBox([
    input_layout,
    widgets.HTML('<hr>'),
    crud_buttons,
    widgets.HBox([btn_clean]),
    widgets.HTML('<hr>'),
    widgets.Label("Log de Operaciones:"),
    output_log
], layout=widgets.Layout(border='solid 1px #A9A9A9', padding='15px', background_color='#EEEEEE', width='520px'))

# Muestra la interfaz
display(app_layout)
```

#Diagrama de clases

```python
# @title
# ============================================================
# DIAGRAMA UML: APLICACIÓN CRUD DE CONTACTOS (PYTHON/IPYWIDGETS)
# ============================================================

# Instalación silenciosa de Graphviz
!apt-get install -y graphviz > /dev/null 2>&1
!pip install graphviz > /dev/null 2>&1

from graphviz import Digraph
from IPython.display import Image, display

# Crear el diagrama
uml = Digraph('CrudAppDiagram', filename='uml_crud_app', format='png')

# --------------------------------------------------------------
# Configuración general y estilo
# --------------------------------------------------------------
uml.attr(
    rankdir='TB', # Top-Bottom (arriba-abajo) para que la InterfazCrud esté arriba
    ranksep='0.8',
    nodesep='0.5',
    bgcolor='#f8f9fa' # Fondo claro
)
uml.attr('graph', dpi='100')

# Estilo base para los nodos de clase
uml.attr('node',
          shape='record',
          style='filled',
          color='#000000',
          fontname='Helvetica',
          fontsize='11')

# --------------------------------------------------------------
# 1. Clase Principal: ContactManagerApp (CON SALTOS DE LÍNEA)
# --------------------------------------------------------------
# Contenido de la clase: Insertamos '\n' para forzar los saltos de línea y compactar el nodo.
main_app_content = '''{
    (C) ContactManagerApp
    |
    + on_create_click(event): void\\l
    + on_read_click(event): void\\l
    + on_delete_click(event): void\\l
    + on_update_click(event): void\\l
    + on_clean_click(event): void\\l
}'''
uml.node('ContactManagerApp',
          main_app_content,
          fillcolor='#DDEEFF', # Azul pálido para la clase principal
          margin='0.15')

# --------------------------------------------------------------
# 2. Clases/Módulos de Operación
# --------------------------------------------------------------
action_color = '#E6FFD8' # Verde muy pálido

# Clase para la operación de Crear
create_content = '''{
    (C) CreateContactOperation
    |
    + execute(data): void
}'''
uml.node('CreateOp', create_content, fillcolor=action_color)

# Clase para la operación de Leer
read_content = '''{
    (C) ReadContactOperation
    |
    + execute(query): data
}'''
uml.node('ReadOp', read_content, fillcolor=action_color)

# Clase para la operación de Actualizar
update_content = '''{
    (C) UpdateContactOperation
    |
    + execute(data): void
}'''
uml.node('UpdateOp', update_content, fillcolor=action_color)

# Clase para la operación de Eliminar
delete_content = '''{
    (C) DeleteContactOperation
    |
    + execute(id): void
}'''
uml.node('DeleteOp', delete_content, fillcolor=action_color)

# --------------------------------------------------------------
# 3. Relaciones (Asociaciones)
# --------------------------------------------------------------
uml.attr('edge', arrowhead='open', style='solid', color='#555555')

# ContactManagerApp -> CreateContactOperation
uml.edge('ContactManagerApp', 'CreateOp', label="usa")

# ContactManagerApp -> ReadContactOperation
uml.edge('ContactManagerApp', 'ReadOp', label="usa")

# ContactManagerApp -> UpdateContactOperation
uml.edge('ContactManagerApp', 'UpdateOp', label="usa")

# ContactManagerApp -> DeleteContactOperation
uml.edge('ContactManagerApp', 'DeleteOp', label="usa")

# --------------------------------------------------------------
# Generar y mostrar el diagrama
# --------------------------------------------------------------
output_path = uml.render(cleanup=True)
display(Image(filename=output_path))
```

```python
# @title
!apt-get install -y graphviz > /dev/null 2>&1
!pip install graphviz > /dev/null 2>&1

from graphviz import Digraph
from IPython.display import Image, display

# Crear el diagrama
dot = Digraph('CasosDeUsoContactos', filename='uml_casos_uso_contactos', format='png')

# --------------------------------------------------------------
# Configuración general y estilo
# --------------------------------------------------------------
dot.attr(
    rankdir='LR',
    bgcolor='#f8f9fa',
    splines='ortho',
    dpi='100',
    nodesep='0.5',
    ranksep='0.7'
)

# --------------------------------------------------------------
# 1. Actor (Stickman compatible con Colab)
# --------------------------------------------------------------
stickman_html = '''<
<TABLE BORDER="0" CELLBORDER="0" CELLPADDING="0">
  <TR><TD ALIGN="CENTER"><FONT FACE="monospace" POINT-SIZE="12">O</FONT></TD></TR>
  <TR><TD ALIGN="CENTER"><FONT FACE="monospace" POINT-SIZE="10">/|\\</FONT></TD></TR>
  <TR><TD ALIGN="CENTER"><FONT FACE="monospace" POINT-SIZE="10">/ \\</FONT></TD></TR>
  <TR><TD ALIGN="CENTER"><FONT POINT-SIZE="10">Usuario</FONT></TD></TR>
</TABLE>>'''

# El nodo del actor es de tipo 'none'
dot.node('actor', stickman_html, shape='none')

# --------------------------------------------------------------
# 2. Casos de Uso (Óvalos CRUD)
# --------------------------------------------------------------
# Estilo para los Casos de Uso (óvalos)
dot.attr('node',
         shape='ellipse',
         style='filled',
         fillcolor='#E0E0E0', # Gris claro
         color='#333333',     # Borde
         fontname='Helvetica',
         fontsize='11',
         margin='0.1')

dot.node('cu_crear', 'Crear contacto', fillcolor='#7abdc2')
dot.node('cu_leer', 'Leer contactos',fillcolor='#7abdc2')
dot.node('cu_actualizar', 'Actualizar contacto',fillcolor='#7abdc2')
dot.node('cu_eliminar', 'Eliminar contacto',fillcolor='#7abdc2')
dot.node('cu_limpiar', 'Limpiar campos', fontsize='10', fillcolor='#96d9d5')


# --------------------------------------------------------------
# 3. Relaciones (Asociaciones)
# --------------------------------------------------------------
dot.attr('edge', arrowhead='normal', style='solid', color='#555555', arrowsize='0.8')

dot.edge('actor', 'cu_crear')
dot.edge('actor', 'cu_leer')
dot.edge('actor', 'cu_actualizar')
dot.edge('actor', 'cu_eliminar')
dot.edge('actor', 'cu_limpiar')


# --------------------------------------------------------------
# Generar y mostrar el diagrama
# --------------------------------------------------------------
output_path = dot.render(cleanup=True)
display(Image(filename=output_path))
```
